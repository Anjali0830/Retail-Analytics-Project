```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"E:\Retail-Analytics-Project\data\raw\raw_sales_data.csv")

# Show first 5 rows
print(df.head())

# Show dataset info
print(df.info())

print("\nColumns:")
print(df.columns.tolist())

print("\nData Types:")
print(df.dtypes)

# Check for missing values
print("Missing Values:")
print(df.isnull().sum())

# Check for duplicate records
duplicates = df.duplicated().sum()
print("Duplicate Records:", duplicates)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Remove null values
df.dropna(inplace=True)

# Convert date column
print(df.columns.tolist())  # Check column names

df['Order_Date'] = pd.to_datetime(
    df['Order_Date'],
    errors='coerce'
)

# Check datatypes
print(df.dtypes)

# Remove invalid rows
df.dropna(inplace=True)

# Summary statistics
print(df.describe())

# Final dataset check
print("Final Shape:")
print(df.shape)

print("\nMissing Values:")
print(df.isnull().sum())

# Save cleaned data
df.to_csv(r"E:\Retail-Analytics-Project\data\cleaned\cleaned_sales_data.csv", index=False)

print("Data cleaned successfully!")
```

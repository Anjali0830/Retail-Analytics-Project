```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"E:\Retail-Analytics-Project\data\raw\raw_sales_data.csv")

# Show first 5 rows
print(df.head())

# Show dataset info
print(df.info())

# Remove duplicates
df.drop_duplicates(inplace=True)

# Remove null values
df.dropna(inplace=True)

# Convert date column
print(df.columns.tolist())  # Check column names

# Remove any accidental spaces around column names 
df.columns = df.columns.str.strip()

# Parse order/date column (handle either "Order Date" or "Order_Date")
if 'Order Date' in df.columns:
    col = 'Order Date'
elif 'Order_Date' in df.columns:
    col = 'Order_Date'
else:
    raise KeyError("No 'Order Date' or 'Order_Date' column found")

# Try parsing with day-first then fallback to month-first for ambiguous cases
# (remove deprecated/unsupported infer_datetime_format argument)
dt = pd.to_datetime(df[col], errors='coerce', dayfirst=True)
if dt.isna().any():
    fallback = pd.to_datetime(df[col], errors='coerce', dayfirst=False, format='mixed')
    dt = dt.fillna(fallback)

# Standardize column name to Order_Date
df['Order_Date'] = dt
if col != 'Order_Date':
    df.drop(columns=[col], inplace=True)

# Check datatypes
print(df.dtypes)

# Save cleaned data
df.to_csv(r"E:\Retail-Analytics-Project\data\cleaned\cleaned_sales_data.csv", index=False)

print("Data cleaned successfully!")
```
